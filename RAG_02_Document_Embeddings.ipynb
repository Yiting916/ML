# Document embeddings

### SentenceTransformer of HuggingFace
https://huggingface.co/sentence-transformers

文件向量化(Document embeddings)就是把文件的內容使用一組Vector來代表，內容相同的兩份文件的vector距離會比較接近\
使用HuggingFace的SentenceTransformer可以將一段文字轉換成Vector

```python
import numpy as np
from sentence_transformers import SentenceTransformer
sentences = ["14縣市大雨特報！雙北慎防大雷雨 新北多處山區溪水暴漲警報",
             "因應鋒面大雨來襲 竹市應變中心強化三級開設嚴陣以待",
             "雨彈開轟！北北基14縣市大雨特報　下到入夜",
             "觀光署宣布：6月起到年底止 花東6路線「台灣好行」免費",
             "〈鴻海股東會〉劉揚偉：GB200供不應求明年AI伺服器業績破兆元 合併營收衝7兆",
             "史上最大獵女巫行動落幕：LayerZero上演幣圈囚徒困境",
             "狩獵進化！《魔物獵人Wilds》首支宣傳影片，新魔物、新荒原、新能力公開",
             "張學友「黃牛票」遭清算釋出 排隊搶購出現「代排」",
             "微軟承認 Windows 再爆更新災情！這次連 Mac 用戶都遭殃"
            ]

model_name = "sentence-transformers/all-MiniLM-L6-v2"

model = SentenceTransformer(model_name)
embeddings = model.encode(sentences)
embeddings.shape
```

```python
# 用來計算兩個vector的距離
def distance(x1, x2):
    d = (x1 - x2)
    d = d*d
    return np.mean(d)
```

```python
print('文件距離0-1', distance(embeddings[0], embeddings[1]))
print('文件距離0-2', distance(embeddings[0], embeddings[2]))
print('文件距離0-3', distance(embeddings[0], embeddings[3]))
print('文件距離0-4', distance(embeddings[0], embeddings[4]))
print('文件距離0-5', distance(embeddings[0], embeddings[5]))
print('文件距離0-6', distance(embeddings[0], embeddings[6]))
print('文件距離0-7', distance(embeddings[0], embeddings[7]))
print('文件距離0-8', distance(embeddings[0], embeddings[8]))
```
